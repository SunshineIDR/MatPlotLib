# Vertical Bar_Colour

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
Beginers = ["Pablo", "Stef", "Collins", "Blake"]
value = [50, 500, 200, 150]
bar_labels = ["red", "blue", "orange", "green"]
bar_colors = ["tab:red", "tab:blue", "tab:orange", "tab:green"]
ax.bar(Beginers, value, label = bar_labels, color = bar_colors)
ax.set_ylabel("Beginners Value")
ax.set_xlabel("Beginners Name")
ax.legend(title = "Legend")
ax.set_title("Beginers Weight")
plt.show()
```

# Bar Label _ Stacking

```python
import matplotlib.pyplot as plt
import numpy as np
Geomatics = ("Pablo", "Sterling", "Nulu")
Gender = {
    "Male": np.array([10,20,50]),
    "Female": np.array([4,12,35])
}
width = 0.4

fig, ax = plt.subplots()
bottom = np.zeros(3)

for Gen, Gender in Gender.items():
    Plot = ax.bar(Geomatics, Gender, width, label = Gen, bottom = bottom)
    ax.bar_label(Plot)
plt.legend()
plt.show()
```

```python
# Using A Callable Method
State = ["Lagos", "Anambra", "Imo", "Enugu"]
Abuja = [234, 167, 189, 130]
fig, ax = plt.subplots()
bar_con = ax.bar(State, Abuja)
ax.set(ylabel = "Distance km/hr", title = "Distance from Abuja to Different States", ylim = (0,250))
ax.bar_label(bar_con, fmt = lambda x:f'{x *1.6:.1f} km/hr')
```

```python
# Using Random
np.random.seed(123456)
Geomatics = ("Pablo", "Sterling", "Nulu")
y_pos = np.arange(len(Geomatics))
height = 3 + 20 * np.random.rand(len(Geomatics))
error = np.random.rand(len(Geomatics))

fig,ax = plt.subplots()
hbar = ax.barh(y_pos, height, xerr = error, align = "center")
ax.set_yticks(y_pos, labels = Geomatics)
ax.invert_yaxis()

plt.show()
```
