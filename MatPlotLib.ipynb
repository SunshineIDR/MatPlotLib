```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([5, 2, 1, 4, 8, 4, 10, 15])
ypoints = np.array([3, 1, 3, 5, 10, 3, 9, 7])

plt.plot(xpoints, ypoints, linestyle = ":", marker = "o", ms = 10, mec = "r", mfc = "r")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([ 2, 2, 13, 13, 2])
ypoints = np.array([10, 20, 20, 10, 10])
rpoints = np.array([7, 25])

font1 = {"family":"arial",  "size": 10, 'color':"red"}
font2 = {"family":"arial",  "size": 20, 'color':"purple"}

plt.plot(xpoints, ypoints, color = "r", marker = "s", mec = "k", mfc = "k")
plt.plot(rpoints, ls = "dashed", color = "k")

plt.xlabel("Eastings", fontdict = font1)
plt.ylabel("Northings", fontdict = font1)
plt.title("UMOREN STERLING",fontdict = font2, loc = "right")
plt.show()
```
